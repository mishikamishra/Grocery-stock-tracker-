```python
# yeah this is my grocery inventory thing for the assignment lol
# it works dont worry prof

inventory = {
    "Milk": {"price": 3.50, "quantity": 50},
    "Bread": {"price": 2.25, "quantity": 5},
    "Eggs": {"price": 4.00, "quantity": 12},
    "Sugar": {"price": 1.50, "quantity": 15}
}

low_threshold = 10
sep = "-" * 55

def add_item():
    print(sep)
    name = input("enter item name: ").strip().title()

    if name in inventory:
        print(name, "already exists dude, use update instead")
        return

    try:
        price = float(input("price: "))
        qty = int(input("how many: "))
    except:
        print("bruh type numbers")
        return

    if price < 0 or qty < 0:
        print("negative?? nah")
        return

    inventory[name] = {"price": price, "quantity": qty}
    print("added", qty, name, "at", price)

def update_stock():
    print(sep)
    name = input("what item: ").strip().title()

    if name not in inventory:
        print("cant find", name)
        return

    print("current:", inventory[name]["quantity"])

    try:
        change = int(input("add or remove how many (can be negative): "))
    except:
        print("not a number smh")
        return

    new_qty = inventory[name]["quantity"] + change

    if new_qty < 0:
        print("not enough stock bro only got", inventory[name]["quantity"])
        return

    inventory[name]["quantity"] = new_qty
    print("ok now its", new_qty)

def show_inventory():
    print(sep)
    if not inventory:
        print("empty lol")
        return

    print(f"{'ITEM':<15} {'PRICE':<10} {'QTY':<6} STATUS")
    print(sep)

    low_count = 0
    for item in sorted(inventory):
        p = inventory[item]["price"]
        q = inventory[item]["quantity"]

        if q < low_threshold:
            status = "LOW"
            low_count += 1
        else:
            status = "ok"

        print(f"{item:<15} ${p:<9.2f} {q:<6} {status}")

    print(sep)
    print("total items:", len(inventory))
    if low_count:
        print(low_count, "things running low btw")

def total_value():
    total = 0
    for stuff in inventory.values():
        total += stuff["price"] * stuff["quantity"]

    print(sep)
    print("total value: $", round(total, 2))
    print(sep)

def menu():
    while True:
        print("\nGROCERY INVENTORY")
        print("1 add item")
        print("2 update stock")
        print("3 show everything")
        print("4 total value")
        print("5 quit")
        print(sep)

        choice = input("pick 1-5: ").strip()

        if choice == "1":
            add_item()
        elif choice == "2":
            update_stock()
        elif choice == "3":
            show_inventory()
        elif choice == "4":
            total_value()
        elif choice == "5":
            print("bye")
            break
        else:
            print("invalid try again")

# run it
menu()
```
